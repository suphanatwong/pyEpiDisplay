# More (harder) stats functions are coming: regression

```python
# logistic regression display to show if certain food items are associated with nausea
from pyepidisplay.logistic_display import logistic_display
df_results_1 = logistic_display('nausea ~ beefcurry + saltegg', outbreak_df_1)
df_results_1
```

```python
# linear regression display to show if certain food items are associated with earlier onset time
from pyepidisplay.regress_display import regress_display
import statsmodels.formula.api as smf
model = smf.ols('onset ~ beefcurry + saltegg', data=outbreak_df_1).fit()
df_results_2 = regress_display(model)
df_results_2
```
